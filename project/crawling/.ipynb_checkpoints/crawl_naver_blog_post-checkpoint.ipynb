```python
#using NAVER search api
#https://developers.naver.com/docs/search/blog/
client_id = "TVg5OnF27E_ggRpHAQxz" #crawling을 위한 naver seaach blog api key
client_secret = "i3eXdbhuCG"

def makedate(string):   #date scheam에 맞추기 위한 method ('YY-MM-DD')
    return string[:4] + '-' + string[4:6] + '-' + string[6:8]

def get_api_result(keyword, display, start): #api 호출 위한 method, return json 객체
    url = "https://openapi.naver.com/v1/search/blog?query=" + keyword \
          + "&display=" + str(display) \
          + "&start=" + str(start) # json 결과    
    result = requests.get(urlparse(url).geturl(),headers = {"X-Naver-Client-Id":client_id,"X-Naver-Client-Secret":client_secret})
    return result.json()

#호출 결과 list로 생성, keyword:검색어/display:한 번 호출시 출력되는 블로그 갯수/start:검색시 출력되는 페이지 번호.(숫자가 작아질수록 최신글 or 관련도가 높은 글 을 의미) 
def make_list(keyword, display, start): #호출 결과 list로 생성, keyword:검색어/display:한 번 호출시 출력되는 블로그 갯수/start:검색시 출력되는 페이지 번호.(숫자가 작아질수록 최신글 or 관련도가 높은 글 을 의미) 
    crawl_list = []                                   
    json_obj = get_api_result(keyword,display,start)
    if 'display' in json_obj.keys():
        search_count = json_obj['display']  
        for i in range(search_count):
            title = json_obj["items"][i]["title"].replace("<b>","").replace("</b>","") #title에서 b태그 삭제
            link = json_obj['items'][i]['link']
            description = json_obj['items'][i]['description'].replace("<b>","").replace("</b>","")
            postdate = json_obj['items'][i]['postdate']
            row = [makedate(postdate),title, link, description]  #한 행
            crawl_list.append(row) 
        return crawl_list
```
