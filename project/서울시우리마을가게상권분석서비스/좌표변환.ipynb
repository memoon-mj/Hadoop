```python
import csv
from pyproj import Proj, transform #좌표 변환 패키지
import numpy as np
import json
#구글 좌표와 매핑하기 위해서 좌표계를 WGS84로 변환한다.
proj_UTMK = Proj(init = 'epsg:5181') #GRS80
proj_WGS84 = Proj(init = 'epsg:4326') #WGS84, 구글지도 좌표계
count = len(jsonobj["features"])  #jsonobj의 총 갯수
header = ['상권_구분_코드','상권_코드','상권_코드_명','상권_구분_코드_명','x','y','x1','y1','WGS84_x','WGS84_y','WGS84_x1','WGS84_y1'] #생성된 csv파일의 row name
f = open('상권_좌표.csv', 'w', encoding ='utf-8')
wr = csv.writer(f)
wr.writerow(header)
for i in range(count):
    len_coord = len(jsonobj['features'][i]['geometry']['coordinates'][0]) #경계 좌표 (x,y)의 갯수
    mid = int(len_coord/2)  #경계 좌표 point 중에서 중앙값 찾기. min값인 위도, 경도를 찾는다.
    상권_구분_코드=jsonobj["features"][i]["properties"]['TRDAR_SE_C']
    상권_코드=jsonobj["features"][i]["properties"]['TRDAR_CD']
    상권_코드_명 = jsonobj["features"][i]["properties"]['TRDAR_CD_N']
    상권_구분_코드_명 =jsonobj["features"][i]["properties"]['TRDAR_SE_1']
    #shpae 파일의 각 상권 경계 파일에서 (x,y), (x1,y1) 두 점의 위도,경도를 선택한다. 
    #각 상권의 경계에 속하는지 확인을 위해 경계 좌표에서 가장 멀리 떨어진 좌표 두 점을 선택한다. 
    if(jsonobj["features"][i]["geometry"]['type'] == 'Polygon'):
        x = jsonobj['features'][i]['geometry']['coordinates'][0][0][0] 
        y = jsonobj['features'][i]['geometry']['coordinates'][0][0][1]
        x1 = jsonobj['features'][i]['geometry']['coordinates'][0][mid][0]
        y1 = jsonobj['features'][i]['geometry']['coordinates'][0][mid][1]
    else:
        x = jsonobj['features'][i]['geometry']['coordinates'][0][0][0][0] 
        y = jsonobj['features'][i]['geometry']['coordinates'][0][0][0][1]
        x1 = jsonobj['features'][i]['geometry']['coordinates'][0][0][mid][0]
        y1 = jsonobj['features'][i]['geometry']['coordinates'][0][0][mid][1]
    WGS84_x,WGS84_y = transform(proj_UTMK, proj_WGS84, x, y)
    WGS84_x1,WGS84_y1 = transform(proj_UTMK, proj_WGS84,x1, y1)
    insert_row = [상권_구분_코드,상권_코드,상권_코드_명,상권_구분_코드_명,x,y,x1,y1,WGS84_x,WGS84_y,WGS84_x1,WGS84_y1]
    wr.writerow(insert_row)
f.close()
```
